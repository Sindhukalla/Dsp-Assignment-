```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,7])
y=np.array([2,4])
plt.title("line graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y)
plt.show()
```

```python
x=np.array([1,7])
y=np.array([2,4])
plt.title("line graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y,"--",color="black")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([3,4,8,7])
y=np.array([6,4,8,6])
col=(["red","blue","pink","yellow"])
plt.bar(x,y,color=col)
plt.title("Bar graph")
plt.show()
```

```python
x=["a","b","c","d"]
y=[10,8,6,4]
plt.xlabel("alphabets")
plt.ylabel("their values")
plt.title("Bar graph")
plt.show(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
plt.pie(y)
plt.show()
```

```python
y=np.array([35,15,50])
plt.pie(y,labels=["A","B","C",])
plt.legend()
plt.show()
```

```python
y=np.array([35,50,15])
plt.pie(y,labels=["A","B","C"])
plt.legend(title="sum")
plt.show()
```

```python
y=np.array([35,50,15])
plt.pie(y,explode=[0,0.2,0])
plt.show()
```

```python
x=np.random.rand(40)
y=np.random.rand(40)
co=np.random.rand(40)
z=np.random.rand(40)
plt.scatter(x,y,s=z,c=co,alpha=0.6)
plt.show()
```

```python
x=np.random.rand(40)
y=np.random.rand(40)
co=np.random.rand(40)
z=np.random.rand(40)
plt.scatter(x,y,c=co,s=z*100,marker="s",edgecolor="red",linewidth=1)
```
